### Lesson learnt

Beware the jupyter is not reactive! If you update a cell, other cells depending on it are **not** automatically re-evaluated.

This can lead to several sneaky, easy to miss traps:
- You have multiple experiments depending on the same parameters, you update the first parameter and rerun the first experiment, but forget to rerun the second (outdated value)
- You redefine a variable / function for a second experiment, then rerun the first experiment again "just in case" before putting the result in the paper. However, now you are running it with a different parameters

These sneaky traps can easily lead to unreproducible results. 

Some mitigations:
- keep it modular! Don't do too many things in one notebook, keep your notebook focused on one task
- Consider reactive notebooks e.g. [marimo](http://marimo.io/)
